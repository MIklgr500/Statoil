```python
import pandas  as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(df, name=''):
    fig, ax = plt.subplots(1,1,figsize=(10,10))
    ax.plot(df['epoch'].values, df['acc'].values, label='train accuracy')
    ax.plot(df['epoch'].values, df['val_acc'].values, label='validation accuracy')
    ax.legend()
    fig.savefig(name+'.png')
    return ax
```
